```python
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
training_data = pd.read_csv('0000000000002747_training_twitter_x_y_train.csv')
```

### I am going to use text column and airline_sentiment for building model

```python
x_train = list(training_data['text'])
y_train = list(training_data['airline_sentiment'])
x_test = list(testing_data['text'])
```

### For training the model better I am going to combine x_train and x_test

```python
tweets_text = x_train + x_test
```

```python
words_list = [word_tokenize(text.lower()) for text in tweets_text]
```

```python
words_list
```

```python
stop = stopwords.words('english')
punctuations = list(string.punctuation) #The punctuations in the string library is of the datatype string, so converting it into list to add it in the stopwords list
stop = stop + punctuations
```

```python
stop
```

### Data Cleaning: Removing stopwords and lemmatize the word

```python
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stop and w.lower()[0] != '/' and w.lower() != 'https' and w.lower() != 'htpp':
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
clean_word_list = [clean_review(word) for word in words_list]
```

```python
#Convert array of words into single text
clean_text = [" ".join(word) for word in clean_word_list]
```

```python
x_train_words = clean_text[0:10980]
x_test_words = clean_text[10980:]
```

```python
count_vec = CountVectorizer(max_features=2000,ngram_range=(1,3),min_df=0.002,max_df=0.95)
x_train_final_features = count_vec.fit_transform(x_train_words)
```

```python
count_vec.get_feature_names()
```

```python
x_test_final_features = count_vec.transform(x_test_words)
```

### Using Random Forest Classifier

```python
rfc_clf = RandomForestClassifier(n_estimators=500)
rfc_clf.fit(x_train_final_features,y_train)
```

```python
y_pred_rfc = rfc_clf.predict(x_test_final_features)
```
